### Compare your response with "correct" response (i.e., make your own cross-tabulation)

```python
data = pd.read_csv('billboard.csv', encoding='latin-1')
```

```python
data.head()
```

```python
len(data)
```

```python
# find total list of unique elements
genres = data.genre.unique()
genres
```

```python
# get actual probability distribution of data
data.genre.value_counts(normalize=True)
```

```python
# Making fake data - your model will create real data!!
import random

my_guesses = random.choices(genres, weights=data.genre.value_counts(normalize=True), k=317)
my_guesses
```

```python
data['PredictedGenre'] = my_guesses
```

```python
data.head()
```

```python
results = pd.crosstab(data['PredictedGenre'], data['genre'])
results
```

```python
# Taking the sum of the diagonal will sum up the "correct" responses:

np.diag(results).sum()
```

```python
# Pandas can give us the sum of the entire table (note this is the same as the length - there is one count for each song)
total = results.sum().sum()
total
```

```python
# "raw" accuracy is the proportion correct out of the total:
np.diag(results).sum()/results.sum().sum()
```

```python
# randomly guessing a genre would give us a 1 in 9 chance of getting it right
1/9
```

Our model is doing better than chance level!! (Lucky, given that this is fake random data!)
