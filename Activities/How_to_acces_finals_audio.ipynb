### Ground Truth Structure:

For the **keyfinding** data, the ground truth can be found in the .csv file "keys_table_updated.csv" and should be accurate (please let me know if you find any errors or missing data.

For the **tempo** data, you will find the ground truth in a text file format: each audio file has a text file (ends with ".bpm") with a matching identifier for the song ID. The text (.bpm) file has a single line inside, with the value of the tempo in BPM rounded to the nearest whole number.

For the **genre** data, it is simply the name of the folder that the audio was read from (also the filename) that gives the "true" label for the genre.

Each of the ground truth information is found in the absolute paths given above.
